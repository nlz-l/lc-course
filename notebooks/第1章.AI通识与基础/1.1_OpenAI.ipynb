# OpenAI 调用大模型

## 0.准备工作

首先，我们需要安装OpenAI的依赖
安装命令：
```shell
uv add openai
```


## 1.初始化客户端

### 1.1.硬编码API_KEY


````python
from openai import OpenAI
# 创建 OpenAI 客户端
client = OpenAI(
    api_key="**",
    base_url="https://api.deepseek.com"
)
````

### 1.2. 从环境变量中获取API_KEY




读取环境变量需要用到python-dotenv，先安装依赖
安装命令：
```shell
uv add python-dotenv
```
再创建.env文件，并写入api_key
```properties
DEEPSEEK_API_KEY=**
```
接着就可以可以从环境变量中获取api_key了

````python
from dotenv import load_dotenv
import os
from openai import OpenAI
load_dotenv()

api_key = os.getenv("DEEPSEEK_API_KEY")
# print(api_key)

client = OpenAI(
    api_key=api_key,
    base_url="https://api.deepseek.com"
)
````

## 2.访问模型

````python
print("🚀 正在调用大模型...")
response = client.chat.completions.create(
    model="deepseek-chat",
    messages=[
        {"role": "system", "content": "你是一名友好的AI助教。"},
        {"role": "user", "content": "你好，你是谁?"}
    ],
    stream=False,
    temperature=0.9,
)
print("🚀 成功调用大模型...")
````

## 3.获取模型响应结果


````python
print(response.model_dump_json(indent=2))
````

````python
print(response.choices[0].message.content)
````
